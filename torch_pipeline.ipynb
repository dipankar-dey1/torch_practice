### Defining the model

```python
class MySimpleNN():
    def __init__(self,X):
        self.weights = torch.rand(X.shape[1],1, dtype= torch.float64, requires_grad=True)
        self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)
    def forward(self, X):
        z= torch.matmul(X, self.weights) + self.bias
        y_pred = torch.sigmoid(z)
        return y_pred
    def loss_function(self, y_pred, y):
        # Clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

        # Calculate loss
        loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
        return loss
    
```

```python
## Important parameters
learning_rate= 0.1
epochs=25
```

```python
#training
model = MySimpleNN(X_train_tensor)
model.weights
for epoch in range(epochs):

     #Forward pass
     y_pred = model.forward(X_train_tensor)
     #print(y_pred)


    #loss calculate
     loss = model.loss_function(y_pred, y_train_tensor)
     
     #backword pass
     loss.backward()


     #parameters update, 
     with torch.no_grad(): # For not tracking gradient here
        model.weights-=learning_rate * model.weights.grad
        model.bias-= learning_rate * model.bias.grad
        
    #zero gradient  # To not accumalate gradient
     model.weights.grad.zero_()
     model.bias.grad.zero_()
    #Print loss
     print(f"Epoch: {epoch+1}, loss: {loss.item()}")
     

```

```python
class MySimpleNN():

  def __init__(self, X):

    self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, X):
    z = torch.matmul(X, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss
```

### Important Parameters

```python
learning_rate = 0.1
epochs = 25
```

### Training Pipeline

```python
# create model
model = MySimpleNN(X_train_tensor)

# define loop
for epoch in range(epochs):

  # forward pass
  y_pred = model.forward(X_train_tensor)

  # loss calculate
  loss = model.loss_function(y_pred, y_train_tensor)

  # backward pass
  loss.backward()

  # parameters update
  with torch.no_grad():  #To not track gradient here
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # zero gradients , to  not accumate gradient
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

#### Simple 3 node in 1st layer and one  node  in second  layer network

```python
#Practice with 3 neuron in first layer and one neuron in second layer
import torch
import torch.nn as nn
class Model(nn.Module):

    def __init__(self, num_features):

        super().__init__()
        self.linear1 = nn.Linear(num_features, 3) #Linear 1 is the name of the  first layer, with 5 input and 3 ouput
        self.relu = nn.ReLU()    #Applying ReLU on it 
        self.linear2 = nn.Linear(3,1)  # Linear 2 is the name of the second layer with 3 input and one output 
        self.sigmoid = nn.Sigmoid()    #Applying Sigmoid on it
    
    def forward(self, features):

        out = self.linear1(features)
        out = self.relu(out)
        out= self.linear2(out)
        out = self.sigmoid(out)
        
        return out
```

```python
class MySimpleNN(nn.Module):

    def __init__(self, num_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_features,3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )

    def forward(self, features):

        out = self.network(features)
        return out 
    
    def loss_function(self, y_pred, y):
        # Clamp predictions to avoid log(0)
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

        # Binary cross entropy manually
        loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
        return loss
```

### Using built in loss function 

```python
class MySimpleNN(nn.Module):

    def __init__(self, num_features):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(num_features,3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )

    def forward(self, features):

        out = self.network(features)
        return out 
    
criterion = nn.BCELoss()

    
```

```python
model = MySimpleNN(X_train_tensor.shape[1])
for epoch in range(epochs):
    # Forward
    y_pred= model(X_train_tensor)
    #Loss
    loss = criterion(y_pred, y_train_tensor.view(-1,1))

    #Backward pass
    loss.backward()

    #parameterss update
    with torch.no_grad():
        for param in model.parameters():
            param-=learning_rate*param.grad
   #Gradient not to accumalate
   # Reset gradients
    model.zero_grad()  
    print(f"Epoch : {epoch+1}, loss: {loss.item()}")
    
```

#### Use build in optimizer

```python
# Definne the model
class MySimpleNN(nn.Module):
    def __init__(self, num_features):
        super().__init__()
        self.network =  nn.Sequential(
            nn.Linear(num_features,3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )
    def forward(self, features):
        out = self.network(features)
        return out

```

```python
criterion = nn.BCELoss()

model = MySimpleNN(X_train_tensor.shape[1])

optimizer=torch.optim.SGD(model.parameters(), lr= learning_rate)

for epoch in range(epochs):

    #Forward
    y_pred = model(X_train_tensor)

    #Loss
    loss = criterion(y_pred, y_train_tensor.view(-1,1))

    #backkward
    loss.backward()

    #paramter update
    optimizer.step()

    #zero grad
    optimizer.zero_grad()
    print(f"Epoch: {epoch+1} , loss: {loss.item()}")
```
